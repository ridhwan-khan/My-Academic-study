```python
input_1=input()

upper=""
lower=""
digit=""
special=""

for i in input_1:
  asi=ord(i)
  if (65<=asi<=90):
    upper+= chr(asi)
  elif (97<=asi<=122):
    lower+= chr(asi)
  elif (48<=asi<=57):
    digit+= chr(asi)
  elif chr(asi) in "@$#":
    special+= chr(asi)

sum=""


print("upper-case:",upper)
print("lower-case:",lower)
print("digits:",digit)
print("special character:",special)

for iter in range(len(upper)):
   new_word=upper[iter]+lower[iter]+digit[iter]+special[iter]
   sum+=new_word

print(sum)
```
